## Part 2 : Data Visualization

#### Below code plots the data that has been munged in prior notebook Various interprations are provided within the comments, for 4 plots The plots have been analysed for these teams - Australia, South Africa, India, New Zealand, Pakistan, West Indies, Sri Lanka, Bangladesh Data has been collected for all Cricket One Day International (ODI) games that have been played since Jan 2005,  involving the above mentioned teams teams 

```python
import pandas as pd
import statsmodels.api as sm
import pylab as pl
import numpy as np
import matplotlib.pyplot as plt
import pylab 
%matplotlib inline
 
# Read the data in
df = pd.read_csv(r"C:\data\Project\Datasets\Data_Regression\ODI_2005_Onwards\Final_Data\Entire_DataSet.csv")
```

## Plot 1 - Game Stats by Team

##### Australia, India, South Africa have won more games than they have lost games.

```python
my_colors='rg'
plot=pd.crosstab(df.Team, df.Result).plot(kind='bar',stacked=False,color=my_colors)
plt.title('Game Stats by Team - 2005 Onwards')
plot.set_xlabel("Teams",fontweight='bold',color='black')
plot.set_ylabel("Number of Games",fontweight='bold',color='black')
plt.yticks(range(0, 150, 25)) 
plt.title("Game Stats by Team - 2005 Onwards",fontweight='bold',color='black')
locs, labels = plt.xticks()
plt.setp(labels, rotation=45)
```

## Plot 2 - Derive stats for each team againt the other qualified teams 

##### Below plot shows that Australia has been strong against most teams, and they have some real competition only from South Africa

```python
ByResult = df.groupby(['Team','Opposition','Result'])
df_cnt=ByResult['Ground'].count()
aus=df_cnt['Australia']
my_colors='rg'
aus_plot = aus.unstack().plot(kind='bar',stacked=False,color=my_colors)
aus_plot.set_xlabel("Opposition Teams",fontweight='bold',color='black')
aus_plot.set_ylabel("Australia",fontweight='bold',color='black')
plt.yticks(range(0, 30, 2)) 
plt.title("Australia against Opposition",fontweight='bold',color='black')
locs, labels = plt.xticks()
plt.setp(labels, rotation=45)
```

## Plot 3 - Stats by Batting Order

##### Teams that have batted 2nd have won slightly more often than the teams that batted 1st, which is quite surprising !

```python
my_colors='rb'
bat_plot=pd.crosstab(df.Bat, df.Result).plot(kind='bar',stacked=False,color=my_colors)
plt.title('Game Stats by Team - 2005 Onwards')
bat_plot.set_xlabel("Batting Order",fontweight='bold',color='black')
bat_plot.set_ylabel("Number of Games",fontweight='bold',color='black')
plt.yticks(range(0, 450, 50)) 
plt.title("Game Stats by Batting Order - 2005 Onwards",fontweight='bold',color='black')
locs, labels = plt.xticks()
plt.setp(labels, rotation=0)
```

## Plot 4 - % Win by Year ( Top 4 teams only)

##### Below plot was applied for the teams that qualified to semi-finals, to see their performance over the past years. South Africa has won all the games they played during 2008, while Australia has been the strongest since 2014. New Zealand has shown tremendous improvement since 2012

```python
fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(111)

x_series=np.array(pd.unique(df_line['Year'].values.ravel()))
y_series_1 = np.array(df_line.query('Team == "Australia"')['Game_Win_Flag']*100)
y_series_2 = np.array(df_line.query('Team == "India"')['Game_Win_Flag']*100)
y_series_3 = np.array(df_line.query('Team == "South Africa"')['Game_Win_Flag']*100)
y_series_4 = np.array(df_line.query('Team == "New Zealand"')['Game_Win_Flag']*100)

pylab.plot(x_series, y_series_1,'-o',color='Red',label='Australia')
pylab.plot(x_series, y_series_2,'-o',color='royalblue',label='India')
pylab.plot(x_series, y_series_3,'-o',color='lime',label='South Africa')
pylab.plot(x_series, y_series_4,'-o',color="tan",label='New Zealand')

locs, labels = plt.xticks()
plt.setp(labels, rotation=0)
plt.grid()

ax1.set_xlabel('Year of Game',fontsize=14,fontweight='bold')
ax1.set_ylabel('% of Wins',fontsize=14,fontweight='bold')
plt.title("% Win by Year",fontsize=14,fontweight='bold')
plt.xticks(range(2005, 2016, 1)) 
plt.yticks(range(0,110,10)) 
pylab.legend(loc='upper right')
```
